Generate 1000 samples from a normal distribution.

```python
import numpy as np 

arr = np.random.normal(loc=1, scale=1, size=1000)
print(arr)
```

```python
arr = np.random.normal(loc=5, scale=2, size=500)
print("Sample values:", arr[:10])
print("Calculated Mean:", np.mean(arr))
print("Calculated Standard Deviation:", np.std(arr))
```
